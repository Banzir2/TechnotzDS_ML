```python
import numpy as np
from sklearn.svm import SVC

# 'Linear' kernel is usually sufficient after LDA
# 'C=1' is a good starting point for regularization
X_train_final = np.load('data/X_train_final.npz')['data']
y_train = np.load('data/y_train.npz')['data']
X_test_final = np.load('data/X_test_final.npz')['data']
y_test = np.load('data/y_test.npz')['data']

svm_clf = SVC(kernel='rbf', C=10.0, gamma='scale')
svm_clf.fit(X_train_final, y_train)
y_pred = svm_clf.predict(X_test_final)

np.savez_compressed('data/y_pred_svm.npz', data=y_pred)

print(f"SVM Accuracy: {svm_clf.score(X_test_final, y_test) * 100:.2f}%")
```
